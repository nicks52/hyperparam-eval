### Initial model setup

Several models will be built, changing parameter values slightly and comparing to the original model output. Once the optimal parameters have been decided, the final model will be produced.

#### Load the required imports

```python
import keras
from keras.datasets import cifar10
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import matplotlib.pyplot as plt
from keras import backend as K
```

#### Load, reshape, and normalize the data

```python
num_classes = 10
img_x, img_y = 32, 32

# The data, split between train and test sets:
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print('x_train shape: {}'.format(x_train.shape))
print('{} train samples'.format(x_train.shape[0]))
print('{} test samples'.format(x_test.shape[0]))

# Normalize the data
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

# Convert class vectors to binary class matricies
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

#### Function to build the model

```python
def build_model(conv_unit_1=32,
                      conv_unit_2=64,
                      kernel_size_1=(3, 3),
                      kernel_size_2=(3, 3),
                      pool_size=(2, 2),
                      batch_norm=False,
                      dense_unit_1=512,
                      dense_unit_2=0):
    """Builds a CNN model.
    
    Arguments:
      conv_unit_1: Integer (default=64), number of output filters in the 
          first two convolutions
      conv_unit_2: Integer (default=64), number of output filters in the 
          second two convolutions
      kernel_size_1: Tuple of 2 integers (default=(3, 3), for the first 
          two convolutions specifying the height and width of the 2D 
          convolution window.
      kernel_size_2: Tuple of 2 integers (default=(3, 3), for the second
          two convolutions specifying the height and width of the 2D 
          convolution window.
      pool_size: Integer (default=(2, 2)), size of the max pooling windows.
      batch_norm: Boolean (default=False), normalizing the data in the 
          batch after the two convolution steps
      dense_unit_1: Positive integer (default=512), dimensionality of the 
          output space for the first dense layer
      dense_unit_2: Positive integer (default=0), dimensionality of the 
          output space for the second dense layer. If 0 then there is 
          only one dense layer.
    
    Returns:
      The CNN model.
    """
    
    model = Sequential()
    model.add(Conv2D(conv_unit_1, kernel_size=kernel_size_1,
                     activation='relu',
                     input_shape=x_train.shape[1:]))
    model.add(Conv2D(conv_unit_1, kernel_size_1, activation='relu'))
    if batch_norm:
        model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=pool_size))
    model.add(Dropout(0.25))

    model.add(Conv2D(conv_unit_2, kernel_size=kernel_size_2,
                     activation='relu'))
    model.add(Conv2D(conv_unit_2, kernel_size_2, activation='relu'))
    if batch_norm:
        model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=pool_size))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(dense_unit_1, activation='relu'))
    if dense_unit_2 > 0:
        model.add(Dropout(0.5))
        model.add(Dense(dense_unit_2, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

#### Function to compile the model

```python
def compile_model(model, optimizer='Adam'):
    """Compiles the already built model.
    
    Arguments:
      optimizer: The optimizer (default='Adam') options are:
          (SGD, RMSprop, Adadelta, Adam)
      
    Returns:
      The compiled model.
    """
    if optimizer == 'SGD':
        opt = keras.optimizers.SGD()
    elif optimizer == 'Adadelta':
        opt = keras.optimizers.Adadelta()
    elif optimizer == 'RMSprop':
        opt = keras.optimizers.RMSprop()
    else:
        opt = keras.optimizers.Adam()
    
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=opt,
                  metrics=['accuracy'])
    
    return model
```

#### Function to train the model

```python
def fit_model(model, batch_size=128, epochs=40, validation_split=0.2):
    """Trains the compiled model.
    
    Arguments:
      model: compiled model returned from the function "compile_model"
      x_train: Input data
      y_train: Target data
      batch_size: Integer (default=128), number of samples per gradient update. 
      epochs: Integer. Integer (default=30), number of epochs to train the model.
      validation_split: Float between 0 and 1. Fraction of the training
          data to be used as validation data.
    
    Returns:
      The history object returned from the keras function model.fit()
    """
    return model.fit(x_train, y_train,
                      batch_size=batch_size,
                      epochs=epochs,
                      verbose=1,
                      validation_split=validation_split,
                      callbacks=None) #[history])
```

#### Function to evaluate the model

```python
def evaluate_model(model, x_test, y_test):
    """
    Prints the test loss and test accuracy of the model.   
    
    This function should only be run on the final model so as 
    not to "taint" the results by using the test data to inform
    model building decisions.
    
    Arguments:
      model: The trained model.
      x_test: The input test data
      y_test: The target test data
      
    Returns:
      Prints the test loss and test accuracy of the model.
    """
    score = model.evaluate(x_test, y_test, verbose=0)
    print('\nTest loss: {}'.format(score[0]))
    print('Test accuracy: {}'.format(score[1]))
    
    return score
```

#### Function to make a comparison plot

```python
def plot_comparison(hist1, model1, hist2, model2):
    """
    Generate two plots to compare two models.
    
    The first plot compares the training accuracy of the two models. And
    the second plot compares the validation accuracy.
    
    Arguments:
      hist1: History object, the history for model1
      model1: String, the model number for hist1 that is being compared
      hist2: History object, the history for model2
      model2: String, the model number for hist2 that is being compared
    
    Results:
      One plot of the training accuracy and one plot of the validation
      accuracy.
    """
    fig, ax = plt.subplots(1, 2, figsize=(15, 7))
    
    ax[0].plot(hist1.history['accuracy'], label = 'Model {} train'.format(model1), color = 'orange')
    ax[0].plot(hist2.history['accuracy'], label = 'Model {} train'.format(model2), color = 'r')
    ax[0].set_title('Training Accuracy for Model {} and Model {}'.format(model1, model2))
    ax[0].set_ylabel('accuracy')
    ax[0].set_xlabel('epoch')
    ax[0].legend()

    ax[1].plot(hist1.history['val_accuracy'], label = 'Model {} validation'.format(model1), color = 'orange')
    ax[1].plot(hist2.history['val_accuracy'], label = 'Model {} validation'.format(model2), color = 'r')
    ax[1].set_title('Validation Accuracy for Model {} and Model {}'.format(model1, model2))
    ax[1].set_ylabel('accuracy')
    ax[1].set_xlabel('epoch')
    ax[1].legend()

    plt.show()
```

### Model 1: 8 layer CNN

Model using kernel size of (3, 3) for all convolutional layers with relu activation functions and has two sets of two convolutional layers. The first set has 32 filters each and the second set has 64 filters each. After each set of convolutional layers is a maxpooling layer with a pool size of (2, 2). After each maxpooling layer is 25% dropout. Following the second dropout is a dense 512 unit layer. Followed by 50% dropout then the final dense layer that uses a softmax activation function.

```python
model1 = build_model(conv_unit_1=32,
                      conv_unit_2=64,
                      kernel_size_1=(3, 3),
                      kernel_size_2=(3, 3),
                      pool_size=(2, 2),
                      batch_norm=False,
                      dense_unit_1=512,
                      dense_unit_2=0)

model1 = compile_model(model1, 'Adam')
```

```python
hist1 = fit_model(model1, batch_size=128, epochs=40, validation_split=0.2)
```

```python
# Print history of model
print('history contains: {}'.format(hist1.history))
```

### Thoughts on model structure and hyperparameters in CNNs

#### Effect of batch size on model performance

An advantage to using smaller batches is that they are noisy and therefore can reduce a model's likelihood for overfitting to the training data. One of the downsides to using smaller batches is that they typically increase the amount of time required to train a model. For example, Model 10 (batch size of 264) had the fastest training time per epoch and Model 11 (batch size of 64) had the longest training time per epoch. 

While having a smaller batch size typically improves model performance, that was not the case in Model 11 and so a batch size of 128 was used. One can see that using a smaller batch size does produce noisier results when looking at the validation accuracy plot for Model 11 vs Model 1. Model 11 has larger fluctuations in validation accuracy.

#### Importance of a fully connected layer between the convolutional outputs and the final softmax layer in a CNN

The benefit of having a fully connected layer between the convolutional outputs and the final softmax layer is that the fully connected layer can aggregate important features that are identified in different convolutional layers. If the fully connected layer was not used between the convolutional outputs and the final softmax layer, then the softmax layer would do a very poor job of classifying the different images because there would not be an easy way for the CNN to learn the connections between important features and the image's classification. 

Based on the results from Model 9, it appears that adding another fully connected layer can further improve model performance though one should be weary of overfitting the model to the training data.

#### Why use categorical_crossentropy as the loss function?

Categorical cross-entropy loss was used when building the CNNs because of the benefits that cross-entropy loss has when evaluating outputs that are probabilities, and because there are 10 categories, or classes, that each image can be classified as for the CIFAR10 dataset. When a model classifies new data, each class is given a percent likelihood that the image belongs to the specific class. The further the percent likelihood is from 1 for the correct class, the larger the cross-entropy loss will be. In doing so, cross-entropy loss can punish incorrect classifications while rewarding correct classifications. While all loss functions punish incorrect classifications, cross-entropy loss punishes an incorrect classification the further the percent likelihood is from 1 exponentially. By having a larger loss value, the smaller the correct classification is, a model is punished much more for very incorrect classifications.

#### Importance of batch normalization in a CNN

Batch normalization is not a required layer in a CNN, but it is often worth implementing because batch normalization often improves the training accuracy of the model. Batch normalization normalizes the output from a previous activation layer and neural networks generally perform better on normalized data. The normalization is performed by subtracting each output by the mean of the batch and then dividing by the standard deviation of the batch. Once all the values in the batch have been normalized, then the following activation function can be used on the data.

#### Importance of dropout in a CNN

Dropout is the technique of randomly setting a fraction of the input values to zero for each pass through the neural network during training. Randomly dropping a fraction of the dataset forces the model to learn more general patterns in the data and the model cannot rely on consecutive layers of the neural network to overfit to the data.

Dropout is not necessary for designing a model, but it is a useful technique to limit overfitting a model to the training data.

#### Impact of Data Augmentation

The ImageDataGenerator function is a process for data augmentation. The input data is provided to the function and then the function will generate more images by manipulating the data in different ways. By producing many output images from one input image, there is more data to train the model on and each image looks entirely new to the model. The function is most commonly used to provide more training data so that more complex models can be created.

The danger of using data augmentation techniques is that the newly generated data may not accurately represent test data. If the generated data does not accurately represent test data, then the model's test accuracy will decrease.
